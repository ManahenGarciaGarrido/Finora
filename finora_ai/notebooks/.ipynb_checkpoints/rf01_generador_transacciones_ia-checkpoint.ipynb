## 👤 2. Derivación del Perfil de Usuario

A partir del `balance_eur` y `months` se infiere un **perfil financiero coherente**:
salario mensual estimado, alquiler, suscripciones y forma de gasto.  
Esto garantiza que los datos generados sean internamente consistentes.

````python
def derivar_perfil(balance_eur: float, months: int) -> dict:
    """
    Deriva un perfil financiero realista desde el saldo objetivo.

    Enfoque budget-first: el salario se calcula para que
        income_total - expenses_total = balance_eur  (exactamente, sin ajuste)

    Fórmula:
        salary = balance_eur / (savings_rate × (salary_count + extra_count × extra_ratio))
    """
    today      = datetime.now()
    salary_day = random.randint(25, 29)
    extra_ratio  = 0.85 + random.random() * 0.30   # paga extra: 85-115 % del salario
    savings_rate = 0.10 + random.random() * 0.15   # tasa de ahorro objetivo: 10-25 %

    # Pre-calcular cuántas nóminas y pagas extra caen en el período generado
    salary_count = 0
    extra_count  = 0
    for m_off in range(months, -1, -1):
        yr = today.year
        mo = today.month - m_off
        while mo <= 0:
            mo += 12; yr -= 1
        midx = mo - 1  # 0-11
        max_d = calendar.monthrange(yr, mo)[1]
        try:
            sal_date = datetime(yr, mo, min(salary_day, max_d))
            if sal_date <= today:
                salary_count += 1
        except ValueError:
            pass
        if midx in (5, 11):
            ex_day = random.randint(14, 18)
            try:
                ex_date = datetime(yr, mo, min(ex_day, max_d))
                if ex_date <= today:
                    extra_count += 1
            except ValueError:
                pass

    # Derivar salario desde el saldo objetivo
    income_multiplier = max(1, salary_count + extra_count * extra_ratio)
    raw_salary        = balance_eur / (savings_rate * income_multiplier)
    salary            = max(1200, min(3500, round(raw_salary / 10) * 10))  # rango €1200-3500, redondeado a decenas

    rent      = round(salary * (0.28 + random.random() * 0.18), -1)
    telecom   = random.choice([32.90, 38.50, 44.90, 55.00])
    elec_base = round(random.uniform(50, 75), 2)
    community = random.choice([65, 75, 85, 95, 110])
    gym_fee   = random.choice([39.90, 42.50, 49.90, 55.00, 0])
    extra_pay = round(salary * extra_ratio, 2)
    has_hbo   = random.random() > 0.55

    return {
        "salary":        salary,
        "salary_day":    salary_day,
        "salary_count":  salary_count,
        "extra_count":   extra_count,
        "extra_ratio":   extra_ratio,
        "savings_rate":  savings_rate,
        "rent":          rent,
        "telecom":       telecom,
        "elec_base":     elec_base,
        "community":     community,
        "gym_fee":       gym_fee,
        "extra_pay":     extra_pay,
        "has_hbo":       has_hbo,
    }

perfil = derivar_perfil(balance_eur, months)

print("Perfil financiero (budget-first):")
print(f"  💰 Sueldo neto mensual : {perfil['salary']:.0f} €  (día {perfil['salary_day']})")
print(f"  📈 Tasa de ahorro obj. : {perfil['savings_rate']*100:.1f}%")
print(f"  🏠 Alquiler mensual    : {perfil['rent']:.0f} €  ({perfil['rent']/perfil['salary']*100:.0f}% del salario)")
print(f"  📱 Vodafone            : {perfil['telecom']:.2f} €/mes")
print(f"  💡 Iberdrola (base)    : {perfil['elec_base']:.2f} €/mes (±30% estacional)")
print(f"  🏢 Comunidad vecinos   : {perfil['community']:.2f} €/mes")
print(f"  🏋️  Gimnasio            : {'%.2f €/mes' % perfil['gym_fee'] if perfil['gym_fee'] else 'No'}")
print(f"  🎁 Paga extra (jun/dic): {perfil['extra_pay']:.2f} €  ({perfil['extra_count']} pagas en {months} meses)")
print(f"  📺 HBO Max             : {'Sí' if perfil['has_hbo'] else 'No'}")
print()
# Verificación anticipada de la aritmética
_total_income_est = perfil['salary'] * perfil['salary_count'] + perfil['extra_pay'] * perfil['extra_count']
print(f"  Ingresos estimados (nóminas + extras): {_total_income_est:,.0f} €")
print(f"  Saldo esperado al {perfil['savings_rate']*100:.1f}%: {_total_income_est * perfil['savings_rate']:,.0f} €  (objetivo: {balance_eur:.0f} €)")
````

## 🔧 5. Generador Estadístico — Algoritmo Budget-First

Cuando Gemini no está disponible, este generador produce un historial igualmente  
rico mediante **reglas estadísticas** con los mismos 45+ comercios y patrones estacionales.

### Algoritmo en 5 pasos

```
1. Generar ingresos (nóminas + pagas extra + freelance)  →  total_income
2. Generar gastos fijos (alquiler, suscripciones, luz)   →  total_fixed
3. Calcular presupuesto variable:
       variable_budget = total_income − balance_eur − total_fixed
4. Generar gastos variables en bruto con el catálogo
5. Escalar: factor = variable_budget / raw_total
            gastos_variables × factor = variable_budget  (exacto)
```

**Resultado por construcción**: `income − fixed − variable = balance_eur` ✅  
Sin ninguna transacción artificial de ajuste. Las proporciones entre categorías se conservan.

````python
def generar_estadistico(balance_eur: float, months: int, perfil: dict) -> list:
    """
    Generador budget-first. Produce transacciones para 'months' meses hacia atrás.

    El saldo cuadra exactamente por construcción matemática:
        total_ingresos − total_gastos_fijos − total_gastos_variables_escalados = balance_eur

    SIN transacciones de ajuste artificiales.
    Devuelve lista de dicts: {date, desc, cat, pm, amount, type}
    """
    today = datetime.now()
    fijos = construir_gastos_fijos(perfil)

    # ─── PASO 1: Generar ingresos y gastos fijos ─────────────────────────────
    income_txs       = []
    fixed_txs        = []
    total_income_eur = 0.0
    total_fixed_eur  = 0.0

    for idx, m_off in enumerate(range(months, -1, -1)):
        yr = today.year
        mo = today.month - m_off
        while mo <= 0:
            mo += 12; yr -= 1
        midx  = mo - 1   # 0-11
        max_d = calendar.monthrange(yr, mo)[1]

        # Nómina con ligera deriva salarial (más antigua = más baja)
        drift    = 1 + idx * 0.0005
        sal_amt  = round(perfil["salary"] * drift, 2)
        sal_day  = min(perfil["salary_day"], max_d)
        sal_date = datetime(yr, mo, sal_day)
        if sal_date <= today:
            total_income_eur += sal_amt
            income_txs.append({
                "date": sal_date.strftime("%Y-%m-%d"), "desc": "Nómina",
                "cat": "Salario", "pm": "bank_transfer",
                "amount": sal_amt, "type": "income"
            })

        # Paga extra (junio=5, diciembre=11)
        if midx in (5, 11):
            ex_day  = min(random.randint(14, 18), max_d)
            ex_date = datetime(yr, mo, ex_day)
            if ex_date <= today:
                total_income_eur += perfil["extra_pay"]
                income_txs.append({
                    "date": ex_date.strftime("%Y-%m-%d"), "desc": "Paga extra",
                    "cat": "Salario", "pm": "bank_transfer",
                    "amount": perfil["extra_pay"], "type": "income"
                })

        # Ingresos variables ocasionales (freelance, bizum recibido)
        if random.random() < 0.25:
            day = random.randint(1, 20)
            amt = round(random.uniform(300, 1500), 2)
            total_income_eur += amt
            income_txs.append({
                "date": datetime(yr, mo, day).strftime("%Y-%m-%d"),
                "desc": "Freelance cliente", "cat": "Freelance",
                "pm": "bank_transfer", "amount": amt, "type": "income"
            })

        # Gastos fijos
        for fe in fijos:
            try:
                d       = min(fe["day"], max_d)
                tx_date = datetime(yr, mo, d)
                if tx_date > today: continue
                variation = 1 + (random.random() * 2 - 1) * fe["var"]
                if fe["desc"] == "Iberdrola" and midx in (10, 11, 0, 1):
                    variation *= 1.35   # calefacción en invierno
                amt = round(max(0.01, fe["base"] * variation), 2)
                total_fixed_eur += amt
                fixed_txs.append({
                    "date": tx_date.strftime("%Y-%m-%d"),
                    "desc": fe["desc"], "cat": fe["cat"], "pm": fe["pm"],
                    "amount": amt, "type": "expense"
                })
            except (ValueError, KeyError):
                pass

    # ─── PASO 2: Calcular presupuesto para gastos variables ──────────────────
    # variable_budget = ingresos_total − saldo_objetivo − gastos_fijos_total
    variable_budget = total_income_eur - balance_eur - total_fixed_eur

    # Si los gastos fijos ya superan el presupuesto, reducir el alquiler
    if variable_budget < 0:
        overage   = -variable_budget
        rent_txs  = [t for t in fixed_txs if t["desc"] == "Alquiler mensual"]
        if rent_txs:
            cut_each = overage / len(rent_txs)
            for t in rent_txs:
                cut = min(cut_each, max(0.0, t["amount"] - 300.0))
                total_fixed_eur  -= cut
                t["amount"]       = round(t["amount"] - cut, 2)
            variable_budget = total_income_eur - balance_eur - total_fixed_eur
        variable_budget = max(0.0, variable_budget)

    # ─── PASO 3: Generar gastos variables en bruto (sin escalar aún) ─────────
    raw_var_txs   = []
    raw_var_total = 0.0

    for m_off in range(months, -1, -1):
        yr = today.year
        mo = today.month - m_off
        while mo <= 0:
            mo += 12; yr -= 1
        midx = mo - 1

        if midx == 11:          seasonal = 1.25   # diciembre (navidad)
        elif midx in (5, 6, 7): seasonal = 1.18   # verano
        elif midx == 0:         seasonal = 0.90   # enero (post-gasto)
        else:                   seasonal = 1.00

        for m in MERCHANTS:
            freq = m["freq"]
            if freq == "weekly":    occ = random.randint(3, 5)
            elif freq == "biweekly": occ = random.randint(1, 3)
            elif freq == "monthly":  occ = 1 if random.random() > 0.20 else 0
            else:                    occ = 1 if random.random() < 0.30 else 0

            lo, hi = m["range"]
            for _ in range(occ):
                day = random.randint(1, 28)
                try:
                    tx_date = datetime(yr, mo, day)
                except ValueError:
                    continue
                if tx_date > today: continue
                amt = round(random.uniform(lo, hi) * seasonal, 2)
                raw_var_total += amt
                raw_var_txs.append({
                    "date": tx_date.strftime("%Y-%m-%d"),
                    "desc": m["desc"], "cat": m["cat"], "pm": m["pm"],
                    "amount": amt, "type": "expense"
                })

    # ─── PASO 4: Escalar gastos variables al presupuesto exacto ──────────────
    # factor = variable_budget / raw_total  →  sum(scaled) = variable_budget
    scale   = variable_budget / raw_var_total if raw_var_total > 0 else 1.0
    var_txs = [
        {**t, "amount": round(max(0.01, t["amount"] * scale), 2)}
        for t in raw_var_txs
    ]

    # Ajuste de céntimos residuales por redondeos (distribuir en las más grandes)
    residual = variable_budget - sum(t["amount"] for t in var_txs)
    if abs(residual) > 0.01 and var_txs:
        by_size = sorted(var_txs, key=lambda t: t["amount"], reverse=True)
        for t in by_size:
            if abs(residual) < 0.01: break
            max_adj = round(t["amount"] * 0.02, 2)
            adj     = min(residual, max_adj) if residual > 0 else max(residual, -max_adj)
            t["amount"] = round(t["amount"] + adj, 2)
            residual   -= adj

    # ─── PASO 5: Combinar (SIN transacción de ajuste artificial) ─────────────
    # Por construcción: total_income − total_fixed − sum(var_txs) = balance_eur
    return income_txs + fixed_txs + var_txs


txs_estadistico = generar_estadistico(balance_eur, months, perfil)

# Verificación inmediata
_inc = sum(t["amount"] for t in txs_estadistico if t["type"] == "income")
_exp = sum(t["amount"] for t in txs_estadistico if t["type"] == "expense")
_net = _inc - _exp
print(f"✅ Generador budget-first: {len(txs_estadistico)} transacciones para {months} meses")
print(f"   Media por mes    : {len(txs_estadistico) / months:.1f} tx/mes")
print(f"   Total ingresos   : {_inc:>10,.2f} €")
print(f"   Total gastos     : {_exp:>10,.2f} €")
print(f"   Saldo resultante : {_net:>10,.2f} €  (objetivo: {balance_eur:.2f} €)")
print(f"   Diferencia       : {abs(_net - balance_eur):>10,.4f} €  {'✅ exacto' if abs(_net - balance_eur) < 0.05 else '⚠️  verificar'}")
````

## 🤖 6. Generación con Gemini 1.5 Flash

Cuando el API key está disponible, Gemini genera el historial completo de una sola llamada.  
El prompt es muy detallado: define el perfil del usuario, los gastos obligatorios,  
los patrones estacionales y la lista de comercios españoles reales.

````python
def construir_prompt(balance_eur: float, months: int, p: dict) -> str:
    """Construye el prompt completo para Gemini."""
    today      = datetime.now()
    start_yr   = today.year
    start_mo   = today.month - months
    while start_mo <= 0: start_mo += 12; start_yr -= 1
    start_date = f"{start_yr:04d}-{start_mo:02d}-01"
    end_date   = today.strftime("%Y-%m-%d")
    gym_line   = f"   • Gimnasio: {p['gym_fee']}€, día 2, expense, direct_debit" if p["gym_fee"] else ""
    elec_max   = round(p["elec_base"] * 1.5, 2)
    elec_min   = round(p["elec_base"] * 0.65, 2)

    return f"""Eres un generador de datos bancarios realistas para Finora (TFG universitario).
Genera un historial de transacciones bancarias COMPLETO y REALISTA.

PERFIL: Sueldo {p['salary']:.0f}€/mes, Alquiler {p['rent']:.0f}€/mes, Madrid, Saldo~{balance_eur:.0f}€
PERÍODO: {start_date} → {end_date} ({months} meses)

FORMATO (array JSON compacto, SIN texto adicional):
[["YYYY-MM-DD","descripción",importe,"expense|income","payment_method"],...]
payment_method: debit_card|credit_card|bank_transfer|direct_debit|bizum

OBLIGATORIO CADA MES:
• Nómina {p['salary']:.0f}€±2%, días 25-29, income, bank_transfer
• Paga extra {p['extra_pay']:.2f}€ SOLO JUNIO y DICIEMBRE días 14-17, income, bank_transfer
• Alquiler {p['rent']:.0f}€, día 1-2, expense, bank_transfer
• Netflix 15.99€ día 5 / Spotify 9.99€ día 8 / Vodafone {p['telecom']:.2f}€±3% día 10
• Iberdrola {p['elec_base']:.2f}€±30% (NOV-FEB hasta {elec_max}€, JUN-SEP desde {elec_min}€) día 15
• Comunidad vecinos {p['community']:.2f}€ día 3
{gym_line}

GASTOS VARIABLES (15-22 por mes):
- Supermercados 3-5×/mes 15-90€: Mercadona, Carrefour, Lidl, Dia, Aldi
- Cafetería/Bar 5-7×/mes 2-10€: Cafetería, Bar La Tasca
- Restaurantes 2-4×/mes 12-45€: Restaurante, McDonald's, Burger King, Telepizza
- Delivery 1-3×/mes 14-38€: Glovo, Just Eat
- Gasolina 2-3×/mes 45-85€: Repsol, BP, Cepsa, Galp
- Ropa 1-2×/mes 20-130€: Zara, H&M, Mango, Pull&Bear, Primark
- Online 2-3×/mes 10-150€: Amazon, El Corte Inglés
- Ocio 1-2×/mes 7-60€: Cines Yelmo, Steam, PlayStation Store
- Salud 1-2×/mes 8-55€: Farmacia García, Farmacia Central

ESTACIONAL:
• DIC: ropa+40%, ElCorteInglés×3-4, Amazon+50%, cenas+30%
• JUL-AGO: gasolina+25%, restaurantes+30%
• ENE: ropa-40%, Amazon+30% (rebajas)
• ESPORÁDICO: MediaMarkt 80-600€ cada 4-8m, IKEA 40-500€ cada 5-9m, Dentista 80-250€ cada 4-6m

Total: {months*14}-{months*24} transacciones. Importes con decimales realistas (34.75€, 67.30€...)
Devuelve SOLO el array JSON: [ ... ]"""

prompt_ejemplo = construir_prompt(balance_eur, months, perfil)
print("Prompt construido:")
print(f"  Longitud: {len(prompt_ejemplo)} caracteres")
print(f"  Tokens estimados: ~{len(prompt_ejemplo) // 4}")
print()
print("─" * 60)
print(prompt_ejemplo[:800] + "\n[...]")
````

````python
def generar_con_gemini(balance_eur: float, months: int, p: dict, model) -> list | None:
    """
    Llama a Gemini y parsea el resultado.
    Devuelve lista de dicts normalizados o None si falla.
    """
    prompt     = construir_prompt(balance_eur, months, p)
    valid_pms  = {"debit_card", "credit_card", "bank_transfer", "direct_debit", "bizum"}

    resp = model.generate_content(prompt,
        generation_config={"temperature": 0.85, "max_output_tokens": 8192})
    raw = resp.text.strip()

    # Limpiar markdown fences si Gemini las añade
    if raw.startswith("```"):
        raw = raw.split("```")[1]
        if raw.lower().startswith("json"): raw = raw[4:]
        raw = raw.strip()
    if raw.endswith("```"): raw = raw[:-3].strip()

    raw_list = json.loads(raw)
    if not isinstance(raw_list, list): return None

    txs = []
    for item in raw_list:
        if not isinstance(item, (list, tuple)) or len(item) < 5: continue
        try:
            date_str = str(item[0]).strip()
            desc     = str(item[1]).strip()[:200]
            amount   = abs(float(item[2]))
            tx_type  = str(item[3]).strip().lower()
            pm       = str(item[4]).strip().lower()
            if tx_type not in ("expense", "income") or amount <= 0: continue
            if pm not in valid_pms: pm = "debit_card"
            datetime.strptime(date_str, "%Y-%m-%d")
            txs.append({"date": date_str, "desc": desc, "cat": "", "pm": pm,
                        "amount": round(amount, 2), "type": tx_type})
        except (ValueError, TypeError, IndexError):
            continue

    if len(txs) < months * 8:
        print(f"⚠️  Solo {len(txs)} txs — por debajo del mínimo ({months*8}). Usando fallback.")
        return None

    return txs


# ── Generar con Gemini o usar estadístico ─────────────────────────────────────
if _gemini_model:
    print("🤖 Llamando a Gemini 1.5 Flash...")
    try:
        txs_gemini = generar_con_gemini(balance_eur, months, perfil, _gemini_model)
        if txs_gemini:
            txs = txs_gemini
            source = "Gemini 1.5 Flash"
            print(f"✅ Gemini generó {len(txs)} transacciones")
        else:
            txs = txs_estadistico
            source = "Fallback estadístico"
    except Exception as e:
        print(f"⚠️  Gemini falló: {e}. Usando fallback estadístico.")
        txs = txs_estadistico
        source = "Fallback estadístico"
else:
    txs = txs_estadistico
    source = "Fallback estadístico"
    print(f"📊 Usando generador estadístico ({len(txs)} transacciones)")

print(f"\n📌 Fuente: {source}")
print(f"   Total transacciones: {len(txs)}")
````

## ✅ 7. Verificación del Saldo

En el enfoque **budget-first**, no existe ninguna transacción de ajuste artificial.  
El saldo cuadra exactamente **por construcción matemática** en el generador estadístico.

Para la ruta **Gemini**, las transacciones las genera la IA con libertad creativa,  
por lo que puede existir un pequeño desfase. En ese caso se corrige **escalando  
proporcionalmente todos los ingresos existentes** (mismo mecanismo que `banks.js`):

```
factor = (total_income + diff) / total_income
→ cada nómina/freelance sube o baja un % uniforme
→ 0 transacciones artificiales añadidas
→ las proporciones entre gastos se preservan intactas
```

Esto elimina la causa raíz de la categoría "Otros gastos" inflada artificialmente.

````python
# ── Verificación y corrección del saldo ──────────────────────────────────────
# El generador estadístico ya cuadra exactamente (budget-first).
# Para Gemini, si existe desfase se corrige escalando ingresos (sin tx artificiales).

total_income_final  = sum(t["amount"] for t in txs if t["type"] == "income")
total_expense_final = sum(t["amount"] for t in txs if t["type"] == "expense")
saldo_final         = total_income_final - total_expense_final
diff                = balance_eur - saldo_final

print(f"Total ingresos     : {total_income_final:>12,.2f} €")
print(f"Total gastos       : {total_expense_final:>12,.2f} €")
print(f"Saldo neto         : {saldo_final:>12,.2f} €")
print(f"Saldo objetivo     : {balance_eur:>12,.2f} €")
print(f"Diferencia         : {diff:>+12,.4f} €")
print()

if abs(diff) > 0.01:
    # Escalar todos los ingresos proporcionalmente (igual que banks.js Tier 1/2)
    incomes   = [t for t in txs if t["type"] == "income"]
    total_inc = sum(t["amount"] for t in incomes)
    if total_inc > 0:
        factor  = max(0.5, min(3.0, (total_inc + diff) / total_inc))
        applied = 0.0
        for t in incomes:
            new_amt    = max(0.01, round(t["amount"] * factor, 2))
            applied   += new_amt - t["amount"]
            t["amount"] = new_amt
        # Ajuste residual de céntimos en el primer ingreso
        residual = diff - applied
        if abs(residual) > 0.001:
            incomes[0]["amount"] = max(0.01, round(incomes[0]["amount"] + residual, 2))

        total_income_final  = sum(t["amount"] for t in txs if t["type"] == "income")
        total_expense_final = sum(t["amount"] for t in txs if t["type"] == "expense")
        saldo_final         = total_income_final - total_expense_final
        print(f"Factor de escala aplicado: ×{factor:.6f}")
        print(f"Saldo tras corrección    : {saldo_final:>12,.2f} €")

diff_final = abs(saldo_final - balance_eur)
print()
print(f"{'✅ SALDO CUADRA' if diff_final < 0.05 else '❌ ERROR — revisar'}"
      f"  (diferencia residual: {diff_final:.4f} €)")
````

````python
gastos_df = df[df["type"] == "expense"].copy()

# Categorías basadas en la descripción (simula la categorización ML)
cat_map = {
    "Mercadona": "Alimentación", "Carrefour": "Alimentación", "Lidl": "Alimentación",
    "Dia": "Alimentación", "Aldi": "Alimentación", "Eroski": "Alimentación",
    "Cafetería": "Alimentación", "Bar": "Alimentación", "Restaurante": "Alimentación",
    "McDonald": "Alimentación", "Burger": "Alimentación", "Telepizza": "Alimentación",
    "Glovo": "Alimentación", "Just Eat": "Alimentación",
    "Repsol": "Transporte", "BP": "Transporte", "Cepsa": "Transporte", "Galp": "Transporte",
    "Renfe": "Transporte", "Metro": "Transporte", "Cabify": "Transporte", "Aparcamiento": "Transporte", "Peaje": "Transporte",
    "Zara": "Ropa", "H&M": "Ropa", "Mango": "Ropa", "Pull": "Ropa", "Bershka": "Ropa",
    "Primark": "Ropa", "Massimo": "Ropa", "Decathlon": "Ropa", "Nike": "Ropa",
    "Cines": "Ocio", "Steam": "Ocio", "PlayStation": "Ocio", "FNAC": "Ocio", "HBO": "Ocio",
    "Netflix": "Ocio", "Spotify": "Ocio",
    "Amazon": "Compras", "Corte Inglés": "Compras", "MediaMarkt": "Compras",
    "IKEA": "Vivienda", "Leroy": "Vivienda", "Alquiler": "Vivienda", "Comunidad": "Vivienda",
    "Farmacia": "Salud", "Dentista": "Salud", "Óptica": "Salud", "Gimnasio": "Salud",
    "Netflix": "Ocio", "Vodafone": "Servicios", "Iberdrola": "Servicios", "Seguro": "Servicios",
    "Apertura": "Otros", "Cargo": "Otros",
}

def categorizar(desc):
    for k, v in cat_map.items():
        if k.lower() in desc.lower(): return v
    return "Otros"

if (gastos_df["cat"] == "").all():
    gastos_df["cat"] = gastos_df["desc"].apply(categorizar)

cat_totals = gastos_df.groupby("cat")["amount"].sum().sort_values(ascending=False)
colors = plt.cm.Set3(np.linspace(0, 1, len(cat_totals)))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Pie chart
wedges, texts, autotexts = ax1.pie(
    cat_totals.values,
    labels=cat_totals.index,
    autopct=lambda p: f"{p:.1f}%" if p > 3 else "",
    colors=colors,
    startangle=90,
)
for at in autotexts: at.set_fontsize(8)
ax1.set_title("Distribución de gastos por categoría", fontweight="bold")

# Bar chart
bars = ax2.barh(cat_totals.index, cat_totals.values, color=colors)
ax2.set_xlabel("Total gastado (€)")
ax2.set_title(f"Total por categoría — {months} meses", fontweight="bold")
for bar, val in zip(bars, cat_totals.values):
    ax2.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f"{val:,.0f}€", va="center", fontsize=8)
ax2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}€"))

plt.tight_layout()
plt.savefig("rf01_categorias.png", dpi=120, bbox_inches="tight")
plt.show()
print("Total gastos por categoría:")
for cat, total in cat_totals.items():
    pct = total / cat_totals.sum() * 100
    print(f"  {cat:<20}: {total:>8,.2f}€  ({pct:.1f}%)")
````

## 💹 11. Ingresos vs Gastos Mensuales — Patrón Estacional

````python
monthly_income  = df[df["type"]=="income"].groupby("month")["amount"].sum()
monthly_expense = df[df["type"]=="expense"].groupby("month")["amount"].sum()
all_months      = sorted(set(monthly_income.index) | set(monthly_expense.index))
inc_vals        = [monthly_income.get(m, 0) for m in all_months]
exp_vals        = [monthly_expense.get(m, 0) for m in all_months]
savings_vals    = [i - e for i, e in zip(inc_vals, exp_vals)]

x = np.arange(len(all_months))
width = 0.35

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 9), sharex=True)

# Barras ingresos / gastos
b1 = ax1.bar(x - width/2, inc_vals, width, label="Ingresos", color="#4CAF50", alpha=0.8)
b2 = ax1.bar(x + width/2, exp_vals, width, label="Gastos",   color="#F44336", alpha=0.8)
ax1.set_ylabel("Euros (€)")
ax1.set_title("Ingresos vs Gastos mensuales", fontweight="bold")
ax1.legend()
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}€"))

# Ahorro mensual
colors_sav = ["#4CAF50" if s >= 0 else "#F44336" for s in savings_vals]
ax2.bar(x, savings_vals, color=colors_sav, alpha=0.85)
ax2.axhline(0, color="black", linewidth=0.8)
ax2.set_ylabel("Ahorro mensual (€)")
ax2.set_title("Ahorro neto mensual", fontweight="bold")
ax2.set_xticks(x)
ax2.set_xticklabels(all_months, rotation=45, ha="right", fontsize=8)
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}€"))

# Anotar meses especiales
for i, m in enumerate(all_months):
    mo = int(m.split("-")[1])
    if mo == 12: ax1.annotate("🎄", (i, max(inc_vals[i], exp_vals[i])+100), ha="center", fontsize=10)
    if mo in (6, 7): ax1.annotate("☀️", (i, max(inc_vals[i], exp_vals[i])+100), ha="center", fontsize=9)

plt.tight_layout()
plt.savefig("rf01_ingresos_gastos.png", dpi=120, bbox_inches="tight")
plt.show()

avg_savings = np.mean(savings_vals)
pos_months  = sum(1 for s in savings_vals if s > 0)
print(f"Ahorro medio mensual : {avg_savings:,.2f}€")
print(f"Meses en positivo    : {pos_months}/{len(savings_vals)}")
print(f"Mejor mes (ahorro)   : {all_months[np.argmax(savings_vals)]} ({max(savings_vals):,.2f}€)")
print(f"Peor mes (gasto)     : {all_months[np.argmin(savings_vals)]} ({min(savings_vals):,.2f}€)")
````

## 📤 14. Salida para la App

Las transacciones se exportan como JSON para ser consumidas por `banks.js`.  
El campo `cents` (en lugar de `amount`) facilita la aritmética entera en el backend Node.js.

````python
# Serializar al formato que espera banks.js
output = {
    "transactions": [
        {
            "date":   row["date"].strftime("%Y-%m-%d"),
            "desc":   row["desc"],
            "cat":    row["cat"],        # '' si no categorizado aún
            "pm":     row["pm"],
            "cents":  int(round(row["amount"] * 100)),
            "type":   row["type"],
        }
        for _, row in df.sort_values("date").iterrows()
    ],
    "source":  source,
    "summary": {
        "total_transactions": len(df),
        "months_covered":     n_months,
        "balance_eur":        round(saldo_final, 2),
        "total_income_eur":   round(total_income_final, 2),
        "total_expense_eur":  round(total_expense_final, 2),
        "avg_tx_per_month":   round(len(df) / n_months, 1),
    }
}

print(f"Output listo: {len(output['transactions'])} transacciones")
print(f"Fuente: {output['source']}")
print(f"\nSummary:")
for k, v in output["summary"].items():
    print(f"  {k:<24}: {v}")

# Mostrar primeras 3 como JSON
print("\nEjemplo (primeras 3 transacciones en formato JSON):")
print(json.dumps(output["transactions"][:3], ensure_ascii=False, indent=2))
````

### Algoritmo budget-first (Tier 3 y notebook)

```
salary = balance_eur / (savings_rate × income_multiplier)

1. Generar ingresos (nóminas + pagas extra + freelance)  →  total_income
2. Generar gastos fijos (alquiler, suscripciones, luz)   →  total_fixed
3. variable_budget = total_income − balance_eur − total_fixed
4. Generar gastos variables en bruto (catálogo de comercios)
5. scale = variable_budget / raw_total
   gastos_variables × scale = variable_budget  ✅ exacto
```

No existe ninguna categoría "Otros gastos" inflada artificialmente.  
Las distribuciones reflejan gasto **real**: Vivienda ~34%, Servicios ~9%, Alimentación ~8%...

### Corrección para ruta Gemini (Tier 1/2)

Cuando Gemini genera transacciones con un pequeño desfase:
- Se escalan **proporcionalmente todos los ingresos** existentes por un factor uniforme
- Ninguna transacción nueva es añadida
- Las proporciones de gastos por categoría se mantienen intactas

### Módulos de IA que se benefician

- **RF-14 Categorización** — cada transacción pasa por el modelo TF-IDF/NB tras la generación
- **RF-22 Predicción** — ≥18 meses garantizan Gradient Boosting como modelo óptimo
- **RF-23 Anomalías** — 15-22 tx/mes dan una línea base estadística sólida (Z-score)
- **RF-24 Suscripciones** — gastos fijos repetidos en el mismo día cada mes
- **RF-21 Recomendaciones** — ingresos y gastos reales para calcular tasa de ahorro
- **RF-29/30 Estadísticas** — 18+ puntos de datos para gráficos de barras y tendencias

### Arquitectura de resiliencia

```
banks.js → Python AI (Gemini)   → ✅ óptimo    (desfase corregido escalando ingresos)
         → Python AI (fallback)  → ✅ bueno     (budget-first, saldo exacto)
         → JS local              → ✅ siempre   (budget-first, saldo exacto)
```

Ningún camino falla: el usuario siempre obtiene una cuenta demo completamente funcional  
con categorías de gastos realistas y sin datos artificiales.
